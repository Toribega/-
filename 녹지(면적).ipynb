```python
# 소계랑 면적만 뽑아오기
graph1 = graph1[['소재지(시군구)별(1)','소재지(시군구)별(2)','2017.1','2018.1','2019.1']]
graph1
```

```python
graph2 = graph1.drop(graph1.index[[0,1,2]])
graph2
```

```python
graph2 = graph2.loc[graph2["소재지(시군구)별(2)"]=='소계']
graph2
```

```python
graph2 = graph2.drop(['소재지(시군구)별(2)'], axis=1)
```

```python
graph2.rename(columns={'2017.1':'2017','2018.1':'2018','2019.1':'2019'}, 
              inplace = True)
graph2
```

```python
from matplotlib import pyplot as plt
import csv
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname = "c:\Windows\Fonts\malgun.ttf").get_name()
rc("font", family=font_name)


fig = plt.figure()
plt.rcParams['figure.figsize'] = (12,12)
list_1 = ["2017","2018","2019"]

axes1 = fig.add_subplot(2, 3, 1)
axes2 = fig.add_subplot(2, 3, 2)
axes3 = fig.add_subplot(2, 3, 3)
axes4 = fig.add_subplot(2, 3, 4)
axes5 = fig.add_subplot(2, 3, 5)
axes6 = fig.add_subplot(2, 3, 6)

y = []
x = ['2017', '2018','2019']

for i in range(len(list_1)):
    y.append(graph2.iloc[0].loc[list_1[i]])

y_1 = []

for i in range(len(list_1)):
    y_1.append(graph2.iloc[1].loc[list_1[i]])

y_2 = []

for i in range(len(list_1)):
    y_2.append(graph2.iloc[2].loc[list_1[i]])
    
y_3 = []

for i in range(len(list_1)):
    y_3.append(graph2.iloc[3].loc[list_1[i]])
               
y_4 = []

for i in range(len(list_1)):
    y_4.append(graph2.iloc[4].loc[list_1[i]])

y_5 = []

for i in range(len(list_1)):
    y_5.append(graph2.iloc[5].loc[list_1[i]])


axes1.plot(x,y, marker = 'o', color = 'red')
axes2.plot(x,y_1, marker = 'o', color = 'blue')
axes3.plot(x,y_2, marker = 'o', color = 'green')
axes4.plot(x,y_3, marker = 'o', color = 'goldenrod')
axes5.plot(x,y_4, marker = 'o', color = 'darkorange')
axes6.plot(x,y_5, marker = 'o', color = 'cyan')

axes1.set_title('서울')
axes2.set_title('부산')
axes3.set_title('대구')
axes4.set_title('인천')
axes5.set_title('광주')
axes6.set_title('대전')

fig.suptitle('녹지 면적')
fig.tight_layout()
```
