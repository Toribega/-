```python
import warnings
warnings.filterwarnings('ignore')
import csv
import pandas as pd
```

```python
def get_status(s):
    if s < 0.031:
        status = "좋음"
    elif s < 0.091:
        status = "보통"
    elif s < 0.151:
        status = "나쁨"
    else:
        status = "아주 나쁨"
    return status
```

```python
graph_2019 = pd.read_csv("D:\pm2021_01\TeamProject\오존대기오염도2019.csv",encoding='euc-kr', engine='python')
graph_2019['2019총평균'] = graph_2019.mean(axis=1)
graph_2019["상태"] = graph_2019["2019총평균"].apply(lambda s: get_status(s))
graph_2019 = graph_2019[['시도별',"2019총평균","상태"]]
graph_2019['위도']= ['37.5640455','35.1646501','35.8799469','37.4647398','35.1768202','36.3732947',
                    '35.5622483','36.3421518','37.5951929','37.8633712','36.6371933','36.5371389',
                    '35.7273991','34.6932557','36.5566132','35.1786319','33.5039166']
graph_2019['경도']= ['126.8339989','128.9317123','128.4266124','126.5342626','126.7737592','127.2487125',
                    '129.2114124','127.2459857','125.9745611','127.1264148','127.4064073','126.2333368',
                    '126.3695257','125.3766365','128.7242988','127.2696711','126.4949803']
graph_2019
```

```python
m_2019 = folium.Map(location = [36.493438, 128.057604], zoom_start = 7)
```

```python
for i in graph_2019.index:
    sub_lat =  graph_2019.loc[i,'위도']
    sub_long = graph_2019.loc[i,'경도']
    
    title = graph_2019.loc[i,'상태']
    pop = graph_2019.loc[i, '시도별']
    
    color = 'blue'
    if graph_2019.loc[i,'상태'] == '보통':
        color = "green"
        
    folium.Marker(
        location = [sub_lat,sub_long], 
        icon=folium.Icon(icon='cloud', color = color), tooltip=title,
                       popup = pop ).add_to(m_2019)

```
