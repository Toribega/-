```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

region = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']
pm10 = np.array([47, 43, 42, 49, 39, 43, 42, 45, 51, 46, 45, 47, 50, 36, 39, 44, 41])

plt.figure(figsize=(10,10))
plt.subplot(2,1,1)

for i in range(17):
  if pm10[i] < 30:
    plt.bar(region[i], pm10[i], color='blue')
  elif 30 <= pm10[i] < 80:
     plt.bar(region[i], pm10[i], color='green')
  elif 80 <= pm10[i] < 150:
     plt.bar(region[i], pm10[i], color='orange')
  elif pm10[i] >= 150:
     plt.bar(region[i], pm10[i], color='red')

plt.title('2016 pm10')
plt.show()
```
