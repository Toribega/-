```python
import csv
import pandas as pd

def pm10(s):
    if s < 30:
        status = "좋음"
    elif s < 80:
        status = "보통"
    elif s < 150:
        status = "나쁨"
    else:
        status = "아주 나쁨"
    return status

graph = pd.read_csv("C:/Users/TeamProject/미세먼지_PM10__월별_도시별_대기오염도_2017.csv",encoding='euc-kr', engine='python')
graph["총평균"] = graph.mean(axis=1)
graph["상태"] = graph["총평균"].apply(lambda s: pm10(s))
graph_2017 = graph[["시도별","총평균","상태"]]
graph_2017
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

region = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']
pm10 = graph_2017["총평균"]

plt.figure(figsize=(10,10))
plt.subplot(2,1,1)

for i in range(17):
  if pm10[i] < 30:
    plt.bar(region[i], pm10[i], color='blue')
  elif 30 <= pm10[i] < 80:
     plt.bar(region[i], pm10[i], color='green')
  elif 80 <= pm10[i] < 150:
     plt.bar(region[i], pm10[i], color='orange')
  elif pm10[i] >= 150:
     plt.bar(region[i], pm10[i], color='red')

plt.title('2017 pm10')
plt.show()
```
