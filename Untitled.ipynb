```python
graph['총평균'] = graph.mean(axis=1)
graph = graph.drop(["평균"],axis = 1)
graph
```

```python
def get_status(s):
    if s < 0.031:
        status = "좋음"
    elif s < 0.091:
        status = "보통"
    elif s < 0.151:
        status = "나쁨"
    else:
        status = "아주 나쁨"
    return status
```

```python
graph["상태"] = graph["총평균"].apply(lambda s: get_status(s))
graph
```
