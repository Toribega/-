```python
import csv
import pandas as pd

def pm10(s):
    if s < 30:
        status = "좋음"
    elif s < 80:
        status = "보통"
    elif s < 150:
        status = "나쁨"
    else:
        status = "아주 나쁨"
    return status

graph = pd.read_csv("C:/Users/TeamProject/미세먼지_PM10__월별_도시별_대기오염도_2016.csv",encoding='euc-kr', engine='python')
graph["총평균"] = graph.mean(axis=1)
graph["상태"] = graph["총평균"].apply(lambda s: pm10(s))
graph_2015 = graph[["시도별","총평균","상태"]]
graph_2015
```
