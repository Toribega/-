```python
import warnings
warnings.filterwarnings('ignore')
import csv
import pandas as pd
```

```python
def get_status(s):
    if s < 16:
        status = "좋음"
    elif s < 36:
        status = "보통"
    elif s < 76:
        status = "나쁨"
    else:
        status = "아주 나쁨"
    return status
```

```python
graph_2019 = pd.read_csv("D:\pm2021_01\TeamProject\초미세먼지2019.csv",encoding='euc-kr', engine='python')
graph_2019['2019총평균'] = graph_2019.mean(axis=1)
graph_2019["상태"] = graph_2019["2019총평균"].apply(lambda s: get_status(s))
graph_2019 = graph_2019[["2019총평균","상태"]]

graph_2019
```

```python
graph_2018 = pd.read_csv("D:\pm2021_01\TeamProject\초미세먼지2018.csv",encoding='euc-kr', engine='python')
graph_2018['2018총평균'] = graph_2018.mean(axis=1)
graph_2018["상태"] = graph_2018["2018총평균"].apply(lambda s: get_status(s))
graph_2018 = graph_2018[["2018총평균","상태"]]
graph_2018
```

```python
graph_2017 = pd.read_csv("D:\pm2021_01\TeamProject\초미세먼지2017.csv",encoding='euc-kr', engine='python')
graph_2017['2017총평균'] = graph_2017.mean(axis=1)
graph_2017["상태"] = graph_2017["2017총평균"].apply(lambda s: get_status(s))
graph_2017 = graph_2017[["2017총평균","상태"]]
graph_2017
```
